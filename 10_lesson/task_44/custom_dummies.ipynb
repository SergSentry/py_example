#### Source data

Задание 44. В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
```

```python
data = pd.DataFrame(data={'whoAmI':lst})
data.head()
```

---

#### Solve

```python
def get_dummies(dataFrame, columnName):
    
    #make sequence for new column 
    def getValue(nameColumn, nameItem):
        return (1 if row[nameColumn] == nameItem else 0 for label, row in data.iterrows())
    
    #calc unique value by column name columnName
    columns =  list(dataFrame[columnName].unique())
    
    newFrame = pd.DataFrame()
    for item in  columns:
        newFrame[item] = pd.Series(data=getValue(columnName, item), name=item)
    
    return newFrame
```

```python
result = get_dummies(data, 'whoAmI')
```
